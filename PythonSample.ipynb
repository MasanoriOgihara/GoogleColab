```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,label='sin(x)')
plt.plot(x,y2,label='cos(x)')
plt.legend()
plt.grid()
plt.grid('minor')
plt.show()
```
